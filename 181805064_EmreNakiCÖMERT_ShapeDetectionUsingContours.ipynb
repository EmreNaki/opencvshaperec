```python
import cv2
import numpy as np

#Start the webcam
cap = cv2.VideoCapture(0)

while True:
    # Read a frame from the camera
    _, frame = cap.read()

    # Convert the frame to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    #Gaussian blur to the grayscale frame
    blur = cv2.GaussianBlur(gray, (5, 5), 0)
    
    #Threshold to Blur
    _, threshold = cv2.threshold(blur, 127, 255, cv2.THRESH_BINARY)

    # Find contours in the threshold image
    contours, _ = cv2.findContours(threshold, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    
    # Iterate over the contours and draw a rectangle around each shape
    for contour in contours:
  
    # cv2.approxPloyDP() function to approximate the shape
        area = cv2.contourArea(contour)
        approx = cv2.approxPolyDP(
              contour, 0.02 * cv2.arcLength(contour, True), True)
        
    #Coordinates of the detected shapes    
        x = approx.ravel()[0]
        y = approx.ravel()[1] 
      
    #drawContours() function
    #area selection otherwise too many shapes
        if area > 400:
            cv2.drawContours(frame, [contour], 0, (0, 0, 255), 5)
         
    #Naming of each shape       
            if len(approx) == 3:
                cv2.putText(frame, 'Triangle', (x, y),
                     cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)
  
            elif len(approx) == 4:
                cv2.putText(frame, 'Rectangle', (x, y),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)
  
            elif len(approx) == 5:
                cv2.putText(frame, 'Pentagon', (x, y),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)
  
            elif len(approx) == 6:
                cv2.putText(frame, 'Hexagon', (x, y),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)
  
            elif 10 < len(approx):
                cv2.putText(frame, 'circle', (x, y),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)
  

        #Show frame
        cv2.imshow('Shapes', frame)
        
    
         # Check if the user pressed the 'q' key
    if cv2.waitKey(1) & 0xFF == ord('q'):
            cap.release()
            cv2.destroyAllWindows()
        
            break
# Release the webcam and close the window






```
